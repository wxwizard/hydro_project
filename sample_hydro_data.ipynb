```python
sample_data = pd.read_excel('Streamflow Data.xlsx')
sample_data
```

```python
data_list = sample_data['ft3/s'].to_list()
data_list = [value/35.3147 for value in data_list]
data_list
```

```python
# data_list = np.array(data_list)
# log_transformed = [np.log10(item) if item != 0 else 0 for item in data_list]
mean = np.mean(data_list)
sd = np.std(data_list)
print(data_list)
print(mean)
print(sd)
```

```python
standardized_values = []
for item in range(len(data_list)):
    standardized_values.append(((data_list[item]-mean)/sd))
print(standardized_values)

```

```python
june_streamflow = standardized_values[5::12]
july_streamflow = standardized_values[6::12]
august_streamflow = standardized_values[7::12]
len(june_streamflow)
```

```python
sf_df = pd.DataFrame([june_streamflow, july_streamflow, august_streamflow])
sf_df = sf_df.T
# sf_df = sf_df.mean(axis=1).to_list()
sf_df
```

```python
summer_averages = []
for val in range(len(july_streamflow)):
    if len(june_streamflow) == len(july_streamflow) == len(august_streamflow):
        summer_averages.append((june_streamflow[val] + july_streamflow[val] + august_streamflow[val])/3)
print(len(summer_averages))
print(seaice_mean)
```

```python
r = np.corrcoef(seaice_mean, summer_averages)
r[0,1]
```
