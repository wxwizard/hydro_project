```python
extent = pd.read_excel('Sea_Ice_Index_Daily_Extent_G02135_v3.0.xlsx', sheet_name=1)
extent
```

```python
extent = extent.T
extent = extent.drop([1978, 1979, 2021, 2022, 'Unnamed: 0', 'Unnamed: 1', '1981-2010 mean', '1981-2010 median'])
extent = extent.astype('float')
extent = extent.interpolate(method='linear', axis=0)
extent
```

```python
print(extent.loc[:,152:243])
```

```python
import numpy as np
extent = extent.values.flatten()
plt.plot(extent)
```

```python
import scipy.signal as ss
seaice_detrended = ss.detrend(extent)

plot = plt.plot(seaice_detrended)
plot
```

```python
from datetime import datetime as dt, timedelta as td
seaice = np.array(seaice_detrended).reshape(41, 366)
seaice = pd.DataFrame(seaice)
seaice.index = np.arange(1980, 2021)
days = 366

date=[]

for day in range(days):
    initial_day = dt(2020,1,1)
    next_day = initial_day + td(days=day)
    next_day = next_day.strftime('%d %b')
    date.append(next_day)
seaice.columns = date
seaice = seaice.mean(axis=1)
print(seaice)
seaice.to_excel('seaice.xlsx')
```
